## Import libraries

```python
import time
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import datetime as dt
```

## get the content of the website


```python
#1166 results
page = requests.get("https://www.spotahome.com/de/s/london--uk/for-rent:apartments/for-rent:studios/bedrooms:1/bedrooms:2/bedrooms:3/bedrooms:3more")
html = page.content
```

## parse the html and save it into a BeautifulSoup instance


```python
bs = BeautifulSoup(html, 'html.parser')
```

## Appartment type (studio, Wohnung)

### housing type

```python
housings = bs.find_all(class_="homecard-content__type__HomecardContent___OmV4c homecard-content__type--rebranding-style__HomecardContent___OmV4c")
housings_lst = (housing.get_text() for housing in housings)
housings_lst
```

```python
housings_lst = [available.strip() for available in housings_lst]
housings_lst[:5]
```

### available from

```python
availables = bs.find_all(class_="homecard-content__available-from__HomecardContent___OmV4c homecard-content__available-from--rebranding-style__HomecardContent___OmV4c")
available_lst = (available.get_text() for available in availables)
available_lst
```

```python
available_lst = [available.strip() for available in available_lst]
available_lst[:5]
```

### description of the appartement

```python
descriptions= bs.find_all(class_="homecard-content__title__HomecardContent___OmV4c homecard-content__title--rebranding-style__HomecardContent___OmV4c")
descriptions_lst = [description.get_text() for description in descriptions]
descriptions_lst
```

```python
descriptions_lst = [description.strip() for description in descriptions_lst]
descriptions_lst[:5]
```

### price

```python
prices = bs.find_all(class_= "price__Price___OmV4c")
prices_lst = [price.get_text() for price in prices]
prices_lst
```

```python
prices_lst = [price.strip() for price in prices_lst]
prices_lst[:5]
```

### price per month etc

```python
prices_month = bs.find_all(class_= "price-monthly__Price___OmV4c price-monthly--rebranding-style__Price___OmV4c")
prices_month_lst = [price_month.get_text() for price_month in prices_month]
prices_month_lst
```

```python
prices_month_lst = [price_month.strip() for price_month in prices_month_lst]
prices_month_lst[:5]
```

```python
prices_month_lst = [price_month.replace('/', '') for price_month in prices_month_lst]
prices_month_lst[:5]
```

### ID

```python
objects_id = bs.find_all(class_='l-list__item')
objects_id_lst = [object_id.get_text() for object_id in objects_id]
objects_id_lst
```

```python
objects_id_lst = [object_id.replace('/', '') for object_id in objects_id_lst]
objects_id_lst[:5]
```

```python
df = pd.DataFrame()
df
```

```python
df['housing_type']=housings_lst
df['available']=available_lst
df['description']=descriptions_lst
df['price']=prices_lst
df['price_per']=prices_month_lst
display(df)
```

```python
today = dt.datetime.today().strftime('%Y-%m-%d %H:%M') # to set the date in the csv filename
df.to_csv('spotahome_{}.csv'.format(today), sep='\t')
```
